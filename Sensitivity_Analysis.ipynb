```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Colombo Inventory with Different Colombo Selling Rates

```python
df = pd.read_excel('Colombo inventory with different colombo selling rates.xlsx',sheet_name = '1')
date = df.iloc[15:116, 1]
colombo_inventory_rate_25 = df.iloc[15:116,2]
colombo_inventory_rate_50 = df.iloc[15:116,4]
colombo_inventory_rate_75 = df.iloc[15:116,6]
colombo_inventory_rate_100 = df.iloc[15:116,8]
colombo_inventory_rate_125 = df.iloc[15:116,10]
colombo_inventory_rate_150 = df.iloc[15:116,12]
colombo_inventory_rate_175 = df.iloc[15:116,14]
colombo_inventory_rate_200 = df.iloc[15:116,16]
```

```python
plt.plot(date,colombo_inventory_rate_25,label = '25',color='green')
plt.plot(date,colombo_inventory_rate_50, label = '50',color='brown')
plt.plot(date,colombo_inventory_rate_75, label = '75',color='orange')
plt.plot(date,colombo_inventory_rate_100, label = '100',color='purple')
plt.plot(date,colombo_inventory_rate_125, label = '125',color='blue')
plt.plot(date,colombo_inventory_rate_150, label = '150',color='black')
plt.plot(date,colombo_inventory_rate_175, label = '175',color='red')
plt.plot(date,colombo_inventory_rate_200, label = '200',color='gray')
plt.title('Colombo inventory with different colombo selling rates')
plt.xlabel('Days')
plt.legend()
plt.ylabel('Colombo Inventory')
plt.show()
```
